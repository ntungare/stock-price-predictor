```python
import os

directory_path = "../data/stocks"
contents = set(os.listdir(directory_path))
print("Total symbols:", len(contents))
```

```python
import random

random.seed(17)

training =  set(random.sample(sorted(contents), int(len(contents)/2)))
testing = contents - training

print("Total training symbols:", len(training))
print("Total testing symbols:", len(testing))
```

```python
import shutil

training_files_path = os.path.join(directory_path, 'training')
os.makedirs(training_files_path, exist_ok=True)
for filename in training:
    shutil.move(os.path.join(directory_path, filename), os.path.join(training_files_path, filename))

testing_files_path = os.path.join(directory_path, 'testing')
os.makedirs(testing_files_path, exist_ok=True)
for filename in testing:
    shutil.move(os.path.join(directory_path, filename), os.path.join(testing_files_path, filename))
```
